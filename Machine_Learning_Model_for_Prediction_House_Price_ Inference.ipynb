# Model Inference

```python
import pickle
import pandas as pd
import numpy as np
import random
```

```python
# Load All Files

with open('all_process.pkl', 'rb') as file_1:
  model_rf = pickle.load(file_1)
```

```python
np.random.seed(21)
n = 100

area = np.random.uniform(1650,16200,n)
bedrooms = np.random.uniform(1,6,n)
bathrooms = np.random.uniform(1,4,n)
stories = np.random.uniform(1,4,n)
parking = np.random.uniform(0,3,n)


guestroom_options = ['no', 'yes']
guestroom = random.choices(guestroom_options, k=n)

airconditioning_options = ['no', 'yes']
airconditioning = random.choices(airconditioning_options, k=n)

prefarea_options = ['no', 'yes']
prefarea = random.choices(prefarea_options, k=n)

furnishingstatus_options = ['unfurnished', 'semi-furnished','furnished']
furnishingstatus = random.choices(furnishingstatus_options, k=n)


# Membuat DataFrame
data_inf = pd.DataFrame({
    'area': area,
    'bedrooms': bedrooms,
    'bathrooms' : bathrooms,
    'stories': stories,
    'parking': parking,
    'guestroom' : guestroom,
    'airconditioning': airconditioning,
    'prefarea': prefarea,
    'furnishingstatus': furnishingstatus,
    
})

data_inf
```

```python
# Predict Inference-Set

prediction_inf = model_rf.predict(data_inf)
prediction_inf
```

```python
# Convert into DataFrame

prediction_inf_df = pd.DataFrame(prediction_inf, columns=['Price - Prediction'])
prediction_inf_df
```

```python
# Merge Inference-Set and Prediction

tabel_inference = pd.concat([data_inf, prediction_inf_df], axis=1)
tabel_inference
```

# Conceptual Problem

1. Jelaskan latar belakang adanya bagging dan cara kerja bagging !

2. Jelaskan perbedaan cara kerja algoritma Random Forest dengan algoritma boosting yang Anda pilih !

3. Jelaskan apa yang dimaksud dengan Cross Validation !

1. Latar belakang bagging adalah untuk meningkatkan ketepatan model dan variansnya. Ini dicapai dengan menggabungkan hasil prediksi dari berbagai model yang dibangun berdasarkan sampel data yang diambil melalui penggantian. Dengan menggunakan teknik penggantian bootstrap, dataset pelatihan dibagi menjadi sejumlah subset yang sama secara acak dalam ukurannya. Setelah itu, setiap subset digunakan untuk melatih model dasar yang independen, seperti pohon keputusan. Pada dataset yang diuji, setiap model dasar melakukan prediksi. Untuk menghasilkan prediksi akhir, hasil prediksi dari model-model dasar digabungkan, biasanya dengan menggunakan rata-rata (dalam regresi) atau mayoritas suara (dalam klasifikasi).

2. Random Forest adalah model set pohon keputusan. Dengan menggunakan sampel acak dengan penggantian, setiap pohon keputusan dibangun secara independen dan dengan sedikit variasi dalam data latihan. Prediksi akhir dibuat dengan menggunakan prediksi mayoritas (klasifikasi) atau rata-rata (regresi). Random Forest dapat menangani overfitting dan bekerja dengan baik dengan berbagai jenis data. Boosting merupakan algoritma pembelajaran yang menggabungkan model yang lemah, biasanya pohon keputusan sederhana, menjadi model yang kuat. Algoritma peningkatan seperti AdaBoost, Gradient Boosting, dan XGBoost. Ini biasanya dilakukan dengan memberi bobot yang berbeda pada setiap sampel data, berkonsentrasi pada sampel yang salah diklasifikasikan oleh model sebelumnya, sehingga model berikutnya akan lebih berkonsentrasi pada sampel tersebut. Ini menghasilkan model yang kinerjanya meningkat secara bertahap hingga mencapai tingkat akurasi yang sangat tinggi. Sementara Boosting membangun model secara berurutan, dengan tujuan untuk memperbaiki kesalahan model sebelumnya, Random Forest membangun model dasar secara independen dan menggabungkan mereka secara paralel. Ini adalah perbedaan utama.

3. Cross validation adalah teknik yang membagi dataset menjadi beberapa subset yang disebut "lipatan", atau folds, dan melatih dan menguji model secara bergantian pada setiap lipatan. Tujuannya adalah untuk mendapatkan estimasi yang lebih baik tentang seberapa baik model akan berfungsi pada data baru. Cross Validation membantu dalam mengidentifikasi masalah seperti overfitting atau underfitting, dan memungkinkan kita untuk menilai seberapa baik model akan berkinerja pada data baru yang belum pernah dilihat.
